```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [0.8, 0.5, 2.5]
Y = [1.2, 0.2, 0.9]
```

```python
def f(w,b,x):
    return 1.0 / (1.0 + np.exp(-(w*x + b)))

def error (w, b) :
    err = 0.0
    for x,y in zip(X,Y) :
        fx = f(w,b,x)
        err += 0.5 * (fx - y) ** 2
    return err
```

```python
def grad_b(w,b,x,y):
    fx = f(w,b,x)
    return (fx - y) * fx * (1 - fx)

def grad_w(w,b,x,y):
    fx = f(w,b,x)
    return (fx - y) * fx * (1 - fx) * x
```

```python
weights = []
biases = []
error_list = []
```

```python
def do_gradient_descent():
    w, b, eta, max_epochs = -2, -2, 1.0, 1000
    for i in range(max_epochs):
        dw, db = 0,0
        for x,y in zip(X, Y) :
            dw = dw + grad_w(w,b,x,y)
            db = db + grad_b(w,b,x,y)
        w = w - eta * dw
        weights.append(w)
        b = b - eta * db
        biases.append(b)
        error_list.append(error(w,b))
```

```python
do_gradient_descent()
```

```python
print(weights)
```

```python
y = np.arange(0,1000)
```

```python
plt.plot(y, weights)
plt.plot(y, biases)
plt.ylabel("weight/bias")
plt.xlabel("Epoch")
plt.legend(['Weight', 'bias'], loc='upper left')
plt.show()
```

```python
plt.plot(y, error_list)
plt.plot(y, [(i,j) for i,j in zip(weights, biases)])
plt.ylabel("Error/X")
plt.xlabel("Epoch")
plt.legend([ 'error', 'Weight', 'bias'], loc='upper left')
plt.show()
```

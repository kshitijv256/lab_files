# Categorical Graph

```python
x = [100,200,300,400]
y = [10,20,30,40]
plt.figure(figsize=(9,3))
plt.subplot(1,3,1)
plt.bar(x,y, color="red")
plt.subplot(1,3,2)
plt.scatter(x,y, color="blue")
plt.subplot(1,3,3)
plt.plot(x,y, color="green")
plt.show()
```
